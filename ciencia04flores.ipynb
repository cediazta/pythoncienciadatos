Vamos a averiguar que podemos hacer si tuvieramos valores nulos dentro de nuestros datos.
Para trabajar con estos valores, se utiliza la libreria numpy

```python
import numpy as np
```

Incluimos valores nulos en las dos primeras filas: np.nan

```python
df.loc[:8, 'longitud_petalo'] = np.nan
df.head()
```

```python
df.info()
```

Los datos nulos son elementos de los mas normal dentro de un DataFrame, datos que no se han rellenado.
Para tratar datos nulos, tenemos un par de opciones:
dropna() --> Elimina la columna/filas con nulos.
axis=0 / axis=1 --> Filas/Columnas

```python
df_limpio = df.dropna()
df_limpio.head()
```

La otra posibilidad esta en poder operar con los datos dentro del DataFrame. 
Un Null es lo mismo aqui que en base de datos, no se tiene en cuenta.
En lugar de limpiar el DataFrame de nulos, podemos modificar los valores nulos tambien.
Es decir, quitar el null y poner un valor por defecto.
Para ello se uiliza fillna(valor).

```python
df['logntud_petalo'] = df['longitud_petalo'].fillna(999)
df
```

Tambien podemos crear agrupaciones para averiguar caracteristicas respecto a una columna.
Esto es lo mismo que un GROUP BY dentro de SQL Server.

```python
df_grupo = df.groupby('clase')['ancho_petalo'].mean()
df_grupo
```

Por supuesto, podemos realizar hasta dibujos!!!
Con la libreria matplotlib.

```python
import matplotlib.pyplot as plt 
df_grupo.plot(kind='pie', y='ancho_petalo', title="Flores") 
```
